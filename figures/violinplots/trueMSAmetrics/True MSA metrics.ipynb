```python
import pandas as pd
import seaborn as sns
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

with open('errorSq') as f:
    data = [list(map(float, col.split())) for col in f.read().split('\n\n')]  
    dataframe = pd.DataFrame(data)
    dataframe = dataframe.T
    
color = {'r','y','g'}
sns.set(rc={'figure.figsize':(12.75,7.5)}, font_scale = 1.5, style= "white")
ax = sns.violinplot(data=dataframe, scale = 'count', palette = color)
ax.set_xticklabels(['','HCV','','','HIV','','','Ebola',''])
ax.set_ylabel('Error Squared')
legend1 = mpatches.Patch(color='y', label='MAFFT')
legend2 = mpatches.Patch(color='r', label='MUSCLE')
legend3 = mpatches.Patch(color='g', label='Clustal Omega')

legend = plt.legend(loc='upper left',handles=[legend1,legend2,legend3])
frame = legend.get_frame()
frame.set_facecolor('white')
frame.set_edgecolor('white')
```
