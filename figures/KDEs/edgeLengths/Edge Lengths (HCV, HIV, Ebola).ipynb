```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

dataset = pd.read_csv("hivLengths2.csv")
ax = sns.kdeplot(data=dataset)    
sns.set(rc={'figure.figsize':(12.75,7.5)}, font_scale = 2, style= "white")
ax.set_ylabel('Kernel Density Estimate')
ax.set_title("HIV")
plt.xlim(-.1,0.55)
```
