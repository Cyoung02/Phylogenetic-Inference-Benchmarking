```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

with open('URF') as f:
    data = [list(map(float, col.split())) for col in f.read().split('\n\n')]  
    dataframe = pd.DataFrame(data)

plt.figure(figsize=(12, 9))
plt.rcParams.update({'font.size': 19})
ax = sns.heatmap(dataframe, annot = True, fmt = '.4f', cmap = "Reds",vmin = 0.05, vmax = 0.015)
ax.set_xticklabels(['FastTree','IQTREE','IQTREE MFP','RAxML','PhyML'])
#plt.yticks(np.arange(4)+0.5,('True MSA','MAFFT','Clustal Omega','MUSCLE'), rotation=0, va="center")
ax.tick_params(left=False)
ax.set_yticklabels([])
ax.set_title('Unweighted Robinson Foulds (HIV)')

print (dataframe)
```
